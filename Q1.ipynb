```python
import numpy as np
```

```python
A =  np.array([[2,0,4,3,0,1,0,2],
[0,2,4,0,2,3,0,0],
[4,0,1,3,0,1,0,1],
[0,1,0,2,0,0,1,0],
[0,0,2,0,0,4,0,0],
[1,1,0,2,0,1,1,3],
[2,1,3,4,0,2,0,2]])
```

```python
A
```

```python
from math import*
def euclidean_distance(x,y):
    return sqrt(sum(pow(a-b,2) for a, b in zip(x, y)))
```

```python
wt  = len(A[:,0])
S= np.zeros([wt, wt])
for i in range(wt):
    for j in range(wt):
        S[i,j]= euclidean_distance(A[i],A[j])
```
